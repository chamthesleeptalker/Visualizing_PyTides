### Station Data Scrape

```python
from bs4 import BeautifulSoup
import urllib2
import requests
import os

os.environ["NO_PROXY"]="https://tidesandcurrents.noaa.gov/stationhome.html"

#main url
url = "https://tidesandcurrents.noaa.gov/stationhome.html"
#url = "http://google.com"
proxies={
    "http":None,
    "https":None
}

#Delaware Bay
staList = ["8557125","8557380","8556198","8555388","8554399","8537121","8537116","8533295","8537101","8537052","8536931","8536915","8536848","8536849","8536581","8536271","8536110","8536021","8535962","8554501","8555461","8536840"]
staDict={}
availStaDict={}

print "Station ID, Established, Date Removed, Longitude, Latitude"

for i in staList:
    elementObj={}
    r = urllib2.urlopen(url+"?id="+i).read()
    soup = BeautifulSoup(r)
    staInfo = soup.find_all("table", class_="table-condensed")
    
    rowData = staInfo[0].find_all('td')

    for l in range(0,(len(rowData)-1),2):

        key_d = (rowData[l].contents)[0]
        try:
            value_d = (rowData[l+1].contents)[0]
        except:
            pass

        key = key_d.strip(":").encode('ascii','ignore')
        value = value_d.encode('utf-8').strip()
        #value= value_d.decode('unicode_escape').encode('utf-8')

        elementObj[key] = value


    staDict[i]=elementObj
    print str(i)+","+str(staDict[i]['Established'])+","+str(staDict[i]['Date Removed'])+","+str(staDict[i]['Longitude'])+","+str(staDict[i]['Latitude'])
    
    if (str(staDict[i]['Date Removed']) != "cham"):
    
        
        #bearing location to decimal degrees
        lat = str(staDict[i]['Latitude']).split(" ")
        lng = str(staDict[i]['Longitude']).split(" ")
        
        if (lat[2] == 'S'):
            lat_deg = float(lat[0].strip("\xc2\xb0"))
            lat_min = float(lat[1].strip("'"))
            
            lat_decDeg = -1*(lat_deg+(lat_min/60.0))
        else:
            lat_deg = float(lat[0].strip("\xc2\xb0"))
            lat_min = float(lat[1].strip("'"))
            
            lat_decDeg = (lat_deg+(lat_min/60.0))
            
        if (lng[2] == 'W'):
            lng_deg = float(lng[0].strip("\xc2\xb0"))
            lng_min = float(lng[1].strip("'"))
            
            lng_decDeg = -1*(lng_deg+(lng_min/60.0))
        else:
            lng_deg = float(lng[0].strip("\xc2\xb0"))
            lng_min = float(lng[1].strip("'"))
            
            lng_decDeg = (lng_deg+(lng_min/60.0))
        
        availStaDict[str(i)] = {"lat":lat_decDeg,"lng":lng_decDeg,"removed":str(staDict[i]['Date Removed'])}
        
```

### Running Pytides Forecast

```python
#!pip install pytides

from datetime import datetime
from pytides.tide import Tide
import numpy as np
import matplotlib.pyplot as plt

predictionsArray=[]

existingStaKeys.remove('8536840');

for j in existingStaKeys:

    heights = []
    t = []
    
    f = open('/home/chm/acads/gme205/scrape/2004_water_level/data_scrape_'+j+'.csv', 'r')
    for i, line in enumerate(f):
        try:
            date = datetime.strptime(" ".join(line.strip().split(",")[:1]), "%Y-%m-%d %H:%M")
            t.append(date)
        except ValueError:
            continue

        heights.append(float(line.strip().split(",")[1]))
        
    f.close()

    #For a quicker decomposition, use hourly readings rather than 6-minutely readings.
    heights = np.array(heights[::10])
    t = np.array(t[::10])

    ##Prepare a list of datetimes, each 6 minutes apart, for a week.
    prediction_t0 = datetime(2014,1,1)
    hours = 0.1*np.arange(7 * 24 * 10)
    times = Tide._times(prediction_t0, hours)

    ##Fit the tidal data to the harmonic model using Pytides
    my_tide = Tide.decompose(heights, t)

    ##Predict the tides using the Pytides model.
    my_prediction = np.array(my_tide.at(times))
    
    predictionsArray.append(my_prediction)

    print "Pytides Prediction for Station "+j+" done!"

    ##Prepare NOAA's results
    noaa_verified = []
    noaa_predicted = []

    predictions = open('/home/chm/acads/gme205/scrape/2014_predictions/data_scrape_'+j+'.csv', 'r')
    for line in predictions:
        try:
            noaa_predicted.append(float(line.split(",")[1]))
        except:
            pass
    predictions.close()
    
    verified = open('/home/chm/acads/gme205/scrape/2014_water_level/data_scrape_'+j+'.csv', 'r')
    for line in verified:
        try:
            noaa_verified.append(float(line.split(",")[1]))
        except:
            pass
    verified.close()

    print "Fetching NOAA predictions and verified readings!"
    noaa_predicted_np =np.array(noaa_predicted[:1680])
    noaa_verified_np = np.array(noaa_verified[:1680])
    
    noaa_predicted_np =np.array(noaa_predicted[:1680])
    noaa_verified_np = np.array(noaa_verified[:1680])
    
    print "Plotting Results!"
    print "\n\n"
    #Plot the results
    plt.figure(figsize=(10,6))
    plt.plot(hours, my_prediction, label="Pytides")
    plt.plot(hours, noaa_predicted_np, label="NOAA Prediction")
    plt.plot(hours, noaa_verified_np, label="NOAA Verified")
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
              ncol=3, fancybox=True, shadow=True)
    plt.title('Comparison of Pytides and NOAA predictions for Station: ' + str(j))
    plt.xlabel('Hours since ' + str(prediction_t0) + '(GMT)')
    plt.ylabel('Metres')
    plt.show()
    
predictionsArray_np=np.array(predictionsArray)
```

### Data Interpolation

```python
# !pip install pykrige
# !pip install cython
#!pip install scipy

from pykrige.ok import OrdinaryKriging
import pykrige.kriging_tools as kt
import numpy as np

latitude = []
longitude = []

for i in existingStations:
    latitude.append(availStaDict[i]["lat"])
    longitude.append(availStaDict[i]["lng"])
    
minmax_lat=[min(latitude),max(latitude)]
minmax_lng=[min(longitude),max(longitude)]
boundBuff = 0.002
res = 0.002

def interpolate(data,minmax_lat,minmax_lng,res,boundBuff):
    gridx = np.arange(minmax_lng[0]-boundBuff, minmax_lng[1]+boundBuff, res)
    gridy = np.arange(minmax_lat[0]-boundBuff, minmax_lat[1]+boundBuff, res)
    # OrdinaryKriging(column1, column2, column3, columnN, varigram_model='linear')
    # generates a numpy array having shape (gridx, gridy)
    #variogram choices linear,power,spherical,gaussian
    ok = OrdinaryKriging(data[:,0], data[:,1], data[:,2],variogram_model='gaussian',
                          verbose=False, enable_plotting=False)
    # produces a numpy array-z and standard dev-ss
    z, ss = ok.execute('grid', gridx, gridy)

    return z
```

### Creating Raster Images

```python
from osgeo import gdal
from osgeo.gdalconst import *
import numpy as np
from osgeo import osr

import matplotlib.pyplot as plt
%matplotlib inline

gdal.AllRegister()

# Calculate the geotransform
new_geo = ( minmax_lng[0], res, 0, minmax_lat[1], 0, -res )

#Fetch new Projection
proj = osr.SpatialReference()
proj.ImportFromEPSG(4326)

#Raster loop range
interpolateRange = predictionsArray_np.shape[1]

#Stack Array Interpolated values
interpolatedStack=[]

#create image
def outputImage(dataArr, name, cols, rows, noBands):
    #create new file
    driver = gdal.GetDriverByName("GTiff")
    driver.Register()
    
    ods = driver.Create(name+'.tiff',rows,cols,noBands,gdal.GDT_Float32)
    
    
    ods.SetGeoTransform(new_geo)
    ods.SetProjection(proj.ExportToWkt())
    
    for i in range(noBands):
        oBand = ods.GetRasterBand(i+1)
        oBand.WriteArray(dataArr[i],0,0)

    print name+": Image done!"

    ods.FlushCache()
    del ods  

print "Interpolation started..."
    
for i in range(interpolateRange):
    data = np.array([[availStaDict[availStaKeys[0]]["lng"],availStaDict[availStaKeys[0]]["lat"],predictionsArray_np[0][i]],
          [availStaDict[availStaKeys[1]]["lng"],availStaDict[availStaKeys[1]]["lat"],predictionsArray_np[1][i]],
          [availStaDict[availStaKeys[2]]["lng"],availStaDict[availStaKeys[2]]["lat"],predictionsArray_np[2][i]]])
    
    interpolatedValues = interpolate(data,minmax_lat,minmax_lng,res,boundBuff)
    
    interpolatedStack.append(interpolatedValues)
    

cols = interpolatedStack[0].shape[0]
rows = interpolatedStack[0].shape[1]

interpolatedStackMax = np.max(interpolatedStack)
interpolatedStackMin = np.min(interpolatedStack)

noBands = int(interpolateRange)

print "Creating output image..."

outputImage(np.array(interpolatedStack),"interpolated_stack", cols, rows, noBands)
```

### Creation of Overlay Image

```python
from PIL import Image
import numpy as np
import matplotlib.cm as cm
import os

classes = 75

steps = (interpolatedStackMax-interpolatedStackMin)/20

def buildClasses(classes,stackMax, stackMin):
    
    classesArr=[]
    steps = (abs(stackMax)+(stackMin))/classes
    classesArr.append(stackMin + steps) 
    
    for i in range(1,classes):
        classesArr.append(classesArr[i-1] + steps)
    
    return classesArr
    

def buildConditions(classesArr,dataArray,stackMax,stackMin):
    
    conditionsArr=[]
    conditionsArr.append(np.logical_and(dataArray >= stackMin,dataArray < classesArr[0] ))
    
    for i in range(len(classesArr)):
        if i == 0:
            conditionsArr.append(np.logical_and(dataArray >= stackMin,dataArray < classesArr[0] ))
        elif i == (len(classesArr)-1):
            conditionsArr.append(np.logical_and(dataArray >= classesArr[i-1], dataArray <= classesArr[i]))
        else:
            conditionsArr.append(np.logical_and(dataArray >= classesArr[i-1], dataArray < classesArr[i]))
    return conditionsArr
            
def buildNormValues(noClasses):
    
    normVal = []
    ini = 1.0/float(noClasses)
    
    for i in range(1,noClasses+1):
        normVal.append(i*ini)
        
    return normVal

def normalizeArray(dataArray,stackMax, stackMin,noClasses):
    
    classesArr = buildClasses(noClasses,stackMax,stackMin)
    conditionsArr = buildConditions(classesArr,dataArray,stackMax,stackMin)
    normValArr = buildNormValues(noClasses)
    
    for i in range(noClasses):
        dataArray[np.where(conditionsArr[i])]=normValArr[i]
        
    final_values = np.multiply(dataArray,mask_inv)
    
    for i in range(noClasses):
        final_values[np.where(final_values == 0)] = 0
    
    return final_values

def colorizedArray(normalizedArray):
    return cm.YlGnBu(normalizedArray)*255
    
def writeImage(colorizedArray,name,folder):
    img = Image.fromarray(np.uint8(colorizedArray))
    img.save(folder+name+'.png')
#     img.save(folder+name+'.png', 'PNG', transparency=0)
    
overlayFolder = '/home/chm/acads/gme205/scrape/img_overlays/'

try:
    os.mkdir(overlayFolder)
except:
    pass


print "Creating overlay images..."
for i in range(interpolateRange):
    writeImage(colorizedArray(normalizeArray(interpolatedStack[i],interpolatedStackMax,interpolatedStackMin,classes)),"imageOvrly_"+(str(i)),overlayFolder)
    
print "Overlay image done!"
```
